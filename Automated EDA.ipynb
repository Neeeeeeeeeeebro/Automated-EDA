```python
## Analysis of high usage air control areas (usg_wac) ##
```

```python
wac_usage = df.groupby('usg_wac')['Total'].mean()

threshold = wac_usage.mean() + wac_usage.std()

high_usage_wac = wac_usage[wac_usage > threshold]

#print("Pneumatic control areas with high utilization rates")
#print(high_usage_wac)
sns.lineplot(x=high_usage_wac,y=threshold,color='orange')
plt.title('Pneumatic control areas with high utilization rates')
plt.xlabel('Total')
plt.ylabel('usg_wac')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
##  Analysis of high usage air control areas (usg_wac) by seaborn ##
```

```python
wac_usage = df.groupby('usg_wac')['Total'].mean()

threshold = wac_usage.mean() + wac_usage.std()

high_usage_wac = wac_usage[wac_usage > threshold]

plt.figure(figsize=(10, 6))
sns.kdeplot(data=high_usage_wac, shade=True, color='g',label='avrage passenger')

plt.title("Distribution of pneumatic control areas with high utilization rates")
plt.xlabel("average passenger")
plt.ylabel("density")
plt.grid()
plt.legend()
plt.show()
```

```python
### 
total_flights = df['Total'].sum()
print("Total Flights:", total_flights)

df['Year'] = pd.to_datetime(df['usg_apt_id']).dt.year

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Year', y='Total', estimator=sum,color='Turquoise')
plt.title('Total Flights Comparison Between Years')
plt.xlabel('Year')
plt.ylabel('Total Flights')
plt.show()
```

```python
### Analysis of trends over time ###
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(day=1))

monthly_flights = df.groupby('Date')['Total'].sum()

plt.figure(figsize=(8, 6))
plt.plot(monthly_flights.index, monthly_flights.values, marker='o',color='r')
plt.xlabel('the date')
plt.ylabel('total trips')
plt.title('Increase or decrease in total trips through time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
